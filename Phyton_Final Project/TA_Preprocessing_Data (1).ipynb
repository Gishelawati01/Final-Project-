```python
### 1. IMPORT LIBRARY PHYTON
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets as datas
```

```python
### 2. MEMBACA DATA
data = pd.read_csv('1_DataSet.csv')
data
```

```python
### 3. MENGHAPUS DATA KOSONG
# Melihat jumlah data kosong untuk setiap kolom
data.isnull().sum()
```

```python
# Menghapus data kosong
data = data.dropna ()
```

```python
# Melihat jumlah data kosong yang terbaru untuk setiap kolom
data.isnull().sum()
```

```python
### 4. Membatasi Nilai Error Parallax < 0.1 dan tidak boleh bernilai negatif
# e_relatif_paralaks < 10%
df = data
df = df.drop(df[(df['e_relatif_paralaks'] <= 0) | (df['e_relatif_paralaks'] >= 0.1)].index)
### 5. Membatasi Nilai RUWE < 1.4
# ruwe < 1.4
df = df.drop(df[(df['RUWE'] >= 1.4) ].index)
### 6. Membatasi Nilai Redshift < 0.1
# e_relatif_z < 10%
df = df.drop(df[(df['e_relatif_z'] <= -0.1) | (df['e_relatif_z'] >= 0.1)].index)
df
```

```python
### 7. Melihat Statistik Setiap Parameter
df.describe()
```

```python
### 8. Menyimpan data frame yang telah dibersihkan ke dalam file CSV
df.to_csv('2_DataClean.csv', index=False)
```

```python
#Menghitung jarak
def jarak(parallax):
  jarak = 1/ parallax
  return jarak
clean['Jarak(pc)'] = jarak(clean['parallax'])
```
